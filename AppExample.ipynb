```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = "https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s={}.html"
```

```python
l = []
```

```python
for page in range(0, 30, 10):
    url = base_url.format(page)
    print(url)
    result = requests.get(url)
    soup = BeautifulSoup(result.content, "html.parser")
    all = soup.find_all("div", {"class":"propertyRow"})
    for item in all:
        d={}
        price = item.find_all("h4", {"class":"propPrice"})[0].text.replace("\n","").replace(" ","")
        d["Price"] = price
        
        try:
            address = item.find_all("span", {"class":"propAddressCollapse"})[0].text
            d["Address"]=address
        except:
            d["Address"]=None
            
        try:
            locality = item.find_all("span", {"class":"propAddressCollapse"})[1].text
            d["Locality"]=locality
        except:
            d["Locality"]=None
            
        try:
            beds = item.find_all("span", {"class":"infoBed"})[0].text
            d["Beds"] = beds
        except:
            d["beds"] = None

        try:
            full_baths = item.find_all("span", {"class":"infoValueFullBath"})[0].text
            d["Full Baths"]=full_baths
        except:
            d["full_baths"]=None

        try:
            half_baths = item.find_all("span", {"class":"infoValueHalfBath"})[0].text
            d["Half Baths"] = half_baths
        except:
            d["Half Baths"] = None

        for column_group in item.find_all("div", {"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span", {"class":"featureGroup"}),column_group.find_all("span", {"class":"featureName"})):
                if "Lot Size" in feature_group.text:
                    d["Lot Size"] = feature_name.text
        l.append(d)
```

```python
import pandas
```

```python
df = pandas.DataFrame(l)
```
